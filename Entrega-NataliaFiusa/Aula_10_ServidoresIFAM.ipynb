## **Desafio Master Blaster**


---

Infelizmente o Governo Federal n√£o tem mais dinheiro para manter tantos docentes na Institui√ß√£o e pretente deixar todas as disciplicinas com 5


*   Se atualmente a diciplina tem mais de 5 servidores, ela ficar√° com apens 5 docentes;
*   Se a diciplina j√° tiver menos que 5 servidores, a quantidade permanecer√° inalterada

Quantos professores precisar√£o ser despedidos?



```python
# Contagem do n√∫mero de servidores por disciplina de ingresso
contagem_disciplina = df_ss['DISCIPLINA INGRESSO'].value_counts()

# Identificar as disciplinas com mais de 5 servidores
disciplinas_excedentes = contagem_disciplina[contagem_disciplina > 5].index.tolist()

# Inicializar o contador de servidores exclu√≠dos
servidores_excluidos = 0

# Iterar sobre as disciplinas excedentes
for disciplina in disciplinas_excedentes:
    # Filtrar o DataFrame para manter apenas as primeiras 5 ocorr√™ncias de cada disciplina excedente
    index_disciplina_excedente = df_ss[df_ss['DISCIPLINA INGRESSO'] == disciplina].index[5:]
    # Verificar se existem servidores a serem exclu√≠dos
    if len(index_disciplina_excedente) > 0:
        df_ss.drop(index_disciplina_excedente, inplace=True)
        # Atualizar o contador de servidores exclu√≠dos
        servidores_excluidos += len(index_disciplina_excedente)

if servidores_excluidos == 0:
    print("N√£o foram encontrados servidores excedentes.")
else:
    print(f"Foram exclu√≠dos {servidores_excluidos} servidores.")
```
