```python
# Importar o TF, o K e suas fun√ß√µes necess√°rias
# import tensorflow as tf
import keras as K

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python

classifier = Sequential ()
```

```python
# 32 features; formato 3x3; Imagens 64x64; Array 3D (RGB).

# Adicionando a Primeira Camada de Convolu√ß√£o
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))

# Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adicionando a Segunda Camada de Convolu√ß√£o
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))

# Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Flattening - Transformando em 1D
classifier.add(Flatten())

# Full connection
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))

# Compilando a rede
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

# Criar o objeto com as regras de pr√©-processamento
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
# Pr√©-processamento das imagens de treino
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/AULA 7/dataset_treino',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
# Executando o treinamento
classifier.fit_generator(training_set,
                         steps_per_epoch = 20,
                         epochs = 10                         )
```

```python
import numpy as np
from keras.preprocessing import image
from IPython.display import Image

# Carregando a imagem de teste
test_image = image.load_img('/content/drive/MyDrive/AULA 7/dataset_treino/dogs/dog.1.jpg', target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)

# Fazendo a previs√£o usando o modelo classifier
result = classifier.predict(test_image)
class_indices = training_set.class_indices

if result[0][0] == 1:
    prediction = 'Cachorro.'
    accuracy = round(result[0][0] * 100, 2)  # A sa√≠da de previs√£o √© uma probabilidade entre 0 e 1
else:
    prediction = 'Gato.'
    accuracy = round((1 - result[0][0]) * 100, 2)  # A sa√≠da de previs√£o √© uma probabilidade entre 0 e 1

# Exibindo a previs√£o e a acur√°cia
print("Previs√£o:", prediction)
print("Acur√°cia:", accuracy, "%.")

# Exibindo a imagem
Image(filename='/content/drive/MyDrive/AULA 7/dataset_treino/dogs/dog.1.jpg')
```
